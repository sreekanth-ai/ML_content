```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(x="timepoint",y="signal",data = fmri,hue="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data = fmri,style="event",hue="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data = fmri,style="event",markers=True,hue="event")
plt.show()
```

BAR PLOT

```python
import pandas as pd
```

```python
iris = pd.read_csv("/content/IRIS.csv")
iris.head()
```

```python
sns.barplot(y='petal_width',x='species',data=iris)
plt.show()
```

```python
sns.barplot(y='petal_width',x='species',data=iris,hue='species')
plt.grid(True)
plt.show()
```

```python
sns.barplot(y='petal_width',x='species',data=iris,palette='Blues_d')
plt.show()
```

```python
sns.barplot(y='petal_width',x='species',data=iris,palette='vlag')
plt.show()
```

```python
sns.barplot(y='petal_width',x='species',data=iris,palette='rocket')
plt.show()
```

```python
sns.barplot(y='petal_width',x='species',data=iris,color='orange')
plt.show()
```

Scatter plot

```python
sns.scatterplot(y='petal_width',x='species',data=iris,hue='petal_length',style='species')
plt.show()
```

```python
sns.scatterplot(y='petal_length',x='sepal_length',data=iris,hue='species',style='species')
plt.show()
```

distrubution graph or histogram

```python
sns.set_theme()
```

```python
diamond = pd.read_csv("/content/diamond.csv.zip")
```

```python
diamond.head()
```

```python
sns.distplot(diamond['price'])
plt.show()
```

```python
sns.distplot(diamond['price'],hist=False)
plt.show()
```

```python
sns.distplot(diamond['price'],color='red',bins=10)
plt.show()
```

```python
sns.distplot(diamond['price'],color='red',bins=100,kde=False)
plt.show()
```

```python
sns.distplot(diamond['price'],color='red',bins=10,vertical=True)
plt.show()
```

JOINTPLOT

```python
sns.jointplot(x='sepal_length',y='petal_length',hue='species',data=iris)
plt.show()
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=iris)
plt.show()
```

```python
sns.jointplot(x='sepal_length',y='petal_length',color='red',kind='reg',data=iris)
plt.show()
```

BOX PLOT

```python
sns.boxplot(y='sepal_length',x='species',data=iris,hue='species')
plt.show()
```

```python
sns.boxplot(y='sepal_length',x='species',palette='Set1',data=iris,hue='species')
plt.show()
```

```python
sns.boxplot(y='sepal_length',x='species',data=iris,hue='species',linewidth=3)
plt.show()
```

```python
sns.boxplot(y='sepal_length',x='species',data=iris,hue='species',order=['Iris-virginica','Iris-setosa','Iris-versicolor'])
plt.show()
```
