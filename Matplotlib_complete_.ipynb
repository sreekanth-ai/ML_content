```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = 4*x
y
```

```python
plot = plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.title("line plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
y1 = 2*x
y1
```

```python
plt.plot(x,y,color="g",linewidth = 2)
plt.plot(x,y1,color='r',linewidth=6)
#plt.grid(True)
plt.show()
```

```python
x1 = np.arange(1,11)
y1 = 5*x1
y2 = 7*x1

plt.subplot(2,1,1)
plt.plot(x1,y1,color='yellow')
plt.subplot(2,1,2)
plt.plot(x1,y2,color='orange')
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [6,9,3,1,0,4,2,10,7]
plt.scatter(x,y,marker="*",color='g',s=100)
plt.xlabel("x values")
plt.ylabel("y avlues")
plt.title("x vs y")
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [6,9,3,1,0,4,2,10,7]
y1 =[5,4,3,2,1,6,7,8,9]
plt.scatter(x,y1,c='orange',s=200)
plt.scatter(x,y,marker="*",color='g',s=100)
plt.xlabel("x values")
plt.ylabel("y avlues")
plt.title("x vs y")
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [6,9,3,1,0,4,2,10,7]
y1 =[5,4,3,2,1,6,7,8,9]

plt.subplot(1,2,1)
plt.scatter(x,y1,c='orange',s=200)
plt.xlabel("x values")
plt.ylabel("y1 avlues")
plt.title("x vs y1")

plt.subplot(1,2,2)
plt.scatter(x,y,marker="*",color='g',s=100)
plt.xlabel("x values")
plt.ylabel("y avlues")
plt.title("x vs y")

plt.show()
```
