```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
canser = pd.read_csv("/content/data.csv")
```

```python
canser.head()
```

```python
canser['diagnosis'].value_counts()
```

```python
label_encoder = LabelEncoder()
```

```python
labels = label_encoder.fit_transform(canser.diagnosis)
```

```python
canser['target'] = labels
```

```python
canser.head()
```

```python
canser['target'].value_counts()
```
