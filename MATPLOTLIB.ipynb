MATPLOTLIB:

Useful for making plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

PLOTTING THE DATA

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sign wave')
plt.show()
```

```python
plt.plot(x,z)
plt.xlabel('angle')
plt.ylabel('cosine value')
plt.title('cosine wave')
plt.show()
```

```python
#PARABOLA
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r*')
plt.show()
```

```python
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x),'r+')
plt.plot(x,np.cos(x),'g-')
plt.show()
```

BAR PLOT

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Hindi','Spanish','Telugu']
people = [100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('languages')
plt.ylabel("No. of people")
plt.title('languages vs No. of people')
plt.show()
```

PIE CHART

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Hindi','Spanish','Telugu']
people = [100,50,150,40,70]
ax.pie(people,labels=languages,autopct='%1.1f')
plt.show()
```

SCATTER PLOT

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

3D SCATTER PLOT

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Greens')
plt.show()
```
